```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('iris')
print(df.shape)
print(df.columns)
print(df.head())
print(df.info())
print(df.describe())

sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()

df.hist(figsize=(10, 8))
plt.show()

sns.boxplot(data=df)
plt.show()
```
